Working in the two-particle basis the wavefunction may be written as:

$$
\ket{\Psi(K)} = \sum_{j,m} \sum_{\alpha, \beta} \mathrm{e}^{\mathrm{i} K (j + \frac{m}{2})} A_{\alpha, \beta} (m) c_{j,\alpha}^\dagger c_{j+m,\beta}^\dagger \ket{0}
$$

where $j = x_1, \; m = x_2 - x_1 $ are relative coordinates such that $j + \frac{m}{2} = R$, the COM of the two-particles. $\alpha,\beta \in \{1,2,..,s\}$ represent sublattice indices.

The Bethe ansatz assumes a form of $A_{\alpha,\beta} (m) = C_{\alpha,\beta} z^{|m|}$ with $ C_{\alpha,\beta} \in \mathbb{C}$, $z \in \mathbb{R}$, $ C_{\alpha,\beta} = A_{\alpha,\beta}(0)$ and $|z| < 1$. From bosonic symmetry we have the additional constraint of $A_{\alpha,\beta} (m) = A_{\beta,\alpha} (-m)$. 

We rewrite the Schrondinger equation as a follows:

$$
T^{0} \ket{\Psi(K)} + T^{+} \ket{\Psi(K)} + T^{-} \ket{\Psi(K)} +U \delta(m) \ket{\Psi(K)} = E \ket{\Psi(K)}
$$

We then get 6 constraints from boundary equations at m = 0:

$$
(E-U)A_{11}(0) = 2 \mathrm{e}^{\mathrm{i} \phi} A_{12}(0) + 2 A_{13}(0) + 2\cos(K/2) \left[A_{12}(-1) + A_{13}(-1)\right] \\
(E-U)A_{22}(0) = 2 \mathrm{e}^{-\mathrm{i} \phi} A_{12}(0) + 2 A_{13}(0) + 2\cos(K/2)A_{12}(-1)\\
(E-U)A_{33}(0) = 2 A_{13}(0) + 2 A_{13}(0) + 2\cos(K/2)A_{13}(-1)\\
EA_{12}(0) = A_{13}(0) + \mathrm{e}^{-\mathrm{i} \phi} A_{11}(0) + \mathrm{e}^{\mathrm{i} \phi} A_{22}(0) + \mathrm{e}^{\mathrm{i} K/2} A_{11}(-1) + \mathrm{e}^{-\mathrm{i} K/2} \left[A_{22}(-1) + A_{23}(-1)\right]\\
EA_{13}(0) = A_{11}(0) + A_{33}(0) + \mathrm{e}^{\mathrm{i} \phi} A_{23}(0) + \mathrm{e}^{\mathrm{i} K/2} A_{11}(-1) + \mathrm{e}^{-\mathrm{i} K/2} \left[A_{23}(-1) + A_{33}(-1)\right] \\
EA_{23}(0) = A_{23}(0) + \mathrm{e}^{-\mathrm{i} \phi} A_{13}(0) + \mathrm{e}^{\mathrm{i} K/2} \left[A_{12}(-1) + A_{23}(-1)\right]\\
$$

And 9 coupled equations for the bulk where $|m| > 0$:
$$
EA_{11}(m) = \mathrm{e}^{\mathrm{i} \phi} \left[A_{12}(m) + A_{21}(m)\right] + A_{13}(m) + A_{31}(m) + \mathrm{e}^{-\mathrm{i} K/2} \left[A_{12}(m-1) + A_{21}(m+1) + A_{13}(m-1) + A_{31}(m+1)\right]\\
EA_{22}(m) = \mathrm{e}^{-\mathrm{i} \phi} \left[A_{12}(m) + A_{21}(m)\right]  + \mathrm{e}^{\mathrm{i} K/2} \left[A_{12}(m-1) + A_{21}(m+1) \right]\\
EA_{33}(m) = A_{13}(m) + A_{31}(m) + \mathrm{e}^{\mathrm{i} K/2} \left[A_{13}(m-1) + A_{31}(m+1) \right]\\
EA_{12}(m) = A_{32}(m) +  \mathrm{e}^{-\mathrm{i} \phi}A_{11}(m) +  \mathrm{e}^{\mathrm{i} \phi}A_{22}(m) + \mathrm{e}^{\mathrm{i} K/2}A_{11}(m+1)  + \mathrm{e}^{-\mathrm{i} K/2} \left[A_{22}(m+1) + A_{32}(m+1) \right]\\
EA_{21}(m) = A_{23}(m) +  \mathrm{e}^{-\mathrm{i} \phi}A_{11}(m) +  \mathrm{e}^{\mathrm{i} \phi}A_{22}(m) + \mathrm{e}^{\mathrm{i} K/2}A_{11}(m-1)  + \mathrm{e}^{-\mathrm{i} K/2} \left[A_{22}(m-1) + A_{32}(m-1) \right]\\
EA_{13}(m) = A_{11}(m) + A_{33}(m) +  \mathrm{e}^{\mathrm{i} \phi}A_{23}(m) + \mathrm{e}^{\mathrm{i} K/2}A_{11}(m+1)  + \mathrm{e}^{-\mathrm{i} K/2} \left[A_{23}(m+1) + A_{33}(m+1) \right]\\
EA_{31}(m) = A_{11}(m) + A_{33}(m) +  \mathrm{e}^{\mathrm{i} \phi}A_{32}(m) + \mathrm{e}^{\mathrm{i} K/2}A_{11}(m-1)  + \mathrm{e}^{-\mathrm{i} K/2} \left[A_{32}(m-1) + A_{33}(m-1) \right]\\
EA_{23}(m) = A_{21}(m) + \mathrm{e}^{-\mathrm{i} \phi}A_{13}(m) + \mathrm{e}^{\mathrm{i} K/2} \left[A_{13}(m-1) + A_{21}(m+1) \right]\\
EA_{32}(m) = A_{12}(m) + \mathrm{e}^{-\mathrm{i} \phi}A_{31}(m) + \mathrm{e}^{\mathrm{i} K/2} \left[A_{31}(m+1) + A_{12}(m-1) \right]
$$


Inserting the ansatz and dividing equations by $z^m$ yields:

Boundary Equations:
$$
(E-U)C_{11}= \left( 2 \mathrm{e}^{\mathrm{i} \phi} C_{12}+ 2 C_{13} \right)+ z^{-1}\left( 2\cos(K/2) \left[C_{12}+ C_{13}\right] \right)   \\
(E-U)C_{22} = \left(2 \mathrm{e}^{-\mathrm{i} \phi} C_{12}+ 2 C_{13} \right)+ z^{-1} \left( 2\cos(K/2)C_{12} \right) \\
(E-U)C_{33} =  \left(2 C_{13}+ 2 C_{13}\right)+ z^{-1} \left( 2\cos(K/2)C_{13} \right)\\
EC_{12}= \left(C_{13}+ \mathrm{e}^{-\mathrm{i} \phi} C_{11}+ \mathrm{e}^{\mathrm{i} \phi} C_{22}\right) + z^{-1} \left( \mathrm{e}^{\mathrm{i} K/2} C_{11} + \mathrm{e}^{-\mathrm{i} K/2} \left[C_{22} + C_{23}\right] \right)\\
EC_{13}= \left(C_{11}+ C_{33} + \mathrm{e}^{\mathrm{i} \phi} C_{23}\right) + z^{-1} \left( \mathrm{e}^{\mathrm{i} K/2} C_{11} + \mathrm{e}^{-\mathrm{i} K/2} \left[C_{23} + C_{33}\right] \right)\\
EC_{23} = \left(C_{23} + \mathrm{e}^{-\mathrm{i} \phi} C_{13}\right)+ z^{-1} \left( \mathrm{e}^{\mathrm{i} K/2} \left[C_{12} + C_{23}\right] \right)\\
$$

Bulk Equations:
$$
EC_{11} = \left( \mathrm{e}^{\mathrm{i} \phi} \left[C_{12} + C_{21}\right] + C_{13} + C_{31} \right) + z^{-1} \left( \mathrm{e}^{-\mathrm{i} K/2} \left[C_{12} + C_{13} \right] \right) +  z\left( \mathrm{e}^{-\mathrm{i} K/2} \left[ C_{31} + C_{21} \right] \right)\\
EC_{22} = \left( \mathrm{e}^{-\mathrm{i} \phi} \left[C_{12} + C_{21}\right] \right)  + z^{-1}\left( \mathrm{e}^{\mathrm{i} K/2} C_{12}  \right) + z \left( \mathrm{e}^{\mathrm{i} K/2}  C_{21}  \right)\\
EC_{33} = \left( C_{13} + C_{31}\right) + z^{-1}\left( \mathrm{e}^{\mathrm{i} K/2} C_{13}  \right) + z \left( \mathrm{e}^{\mathrm{i} K/2}  C_{31}  \right)\\

EC_{12} = \left(C_{32} +  \mathrm{e}^{-\mathrm{i} \phi}C_{11} +  \mathrm{e}^{\mathrm{i} \phi}C_{22}\right) + z \left( \mathrm{e}^{\mathrm{i} K/2}C_{11}  + \mathrm{e}^{-\mathrm{i} K/2} \left[C_{22} + C_{32} \right] \right)\\
EC_{21} = \left(C_{23} +  \mathrm{e}^{-\mathrm{i} \phi}C_{11} +  \mathrm{e}^{\mathrm{i} \phi}C_{22}\right) + z^{-1} \left( \mathrm{e}^{\mathrm{i} K/2}C_{11}  + \mathrm{e}^{-\mathrm{i} K/2} \left[C_{22} + C_{32} \right] \right)\\
EC_{13} = \left(C_{11} + C_{33} +  \mathrm{e}^{\mathrm{i} \phi}C_{23}\right) + z \left( \mathrm{e}^{\mathrm{i} K/2}C_{11}  + \mathrm{e}^{-\mathrm{i} K/2} \left[C_{23} + C_{33} \right] \right)\\
EC_{31} = \left(C_{11} + C_{33} +  \mathrm{e}^{\mathrm{i} \phi}C_{32}\right) + z^{-1} \left( \mathrm{e}^{\mathrm{i} K/2}C_{11} + \mathrm{e}^{-\mathrm{i} K/2} \left[C_{32} + C_{33} \right] \right)\\
EC_{23} = \left(C_{21} + \mathrm{e}^{-\mathrm{i} \phi}C_{13}\right) + z^{-1} \left( \mathrm{e}^{\mathrm{i} K/2} C_{13}  \right) + z \left( \mathrm{e}^{\mathrm{i} K/2} C_{21} \right)\\
EC_{32} = \left(C_{12} + \mathrm{e}^{-\mathrm{i} \phi}C_{31}\right) + z^{-1} \left( \mathrm{e}^{\mathrm{i} K/2} C_{12}  \right) + z \left( \mathrm{e}^{\mathrm{i} K/2} C_{31} \right)
$$


```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Two–particle dimer dispersion by simultaneous solution of
    det B(E,z) = det M(E,z) = 0
where
  B – 9×9 bulk matrix  (Eqns (30) in the note below)
  M – 9×9 boundary+symmetry matrix (Eqns (24)+(sym) below)

author: (c)2025  Ciro et al.
"""

import numpy as np
from numpy import cos, exp
from scipy.optimize import least_squares
import matplotlib.pyplot as plt


# ----------------------------------------------------------------------
# helper: build the 9×9 BULK matrix B(E,z;K,phi)
# ordering  C = [C11,C22,C33,C12,C21,C13,C31,C23,C32]
# ----------------------------------------------------------------------
def bulk_matrix(E, z, K, phi):
    e_ip,  e_imp  = exp(1j*phi),  exp(-1j*phi)
    e_iK2, e_mK2  = exp(1j*K/2),  exp(-1j*K/2)
    zinv          = 1/z

    B = np.zeros((9, 9), dtype=complex)

    # (1)  E C11 − e^{iφ}(C12+C21) − (C13+C31)
    #      − z^{-1} e^{-iK/2}(C12+C13) − z e^{-iK/2}(C31+C21) = 0
    B[0, 0] = E
    B[0, 3] = -e_ip     - zinv*e_mK2
    B[0, 4] = -e_ip                 - z*e_mK2
    B[0, 5] = -1        - zinv*e_mK2
    B[0, 6] = -1                     - z*e_mK2

    # (2)  E C22 − e^{-iφ}(C12+C21) − z^{-1}e^{iK/2}C12 − z e^{iK/2}C21 = 0
    B[1, 1] = E
    B[1, 3] = -e_imp   - zinv*e_iK2
    B[1, 4] = -e_imp                - z*e_iK2

    # (3)  E C33 − (C13+C31) − z^{-1}e^{iK/2}C13 − z e^{iK/2}C31 = 0
    B[2, 2] = E
    B[2, 5] = -1      - zinv*e_iK2
    B[2, 6] = -1                   - z*e_iK2

    # (4)  E C12 − C32 − e^{-iφ}C11 − e^{iφ}C22
    #      − z(e^{iK/2}C11 + e^{-iK/2}(C22+C32)) = 0
    B[3, 0] = -e_imp   - z*e_iK2
    B[3, 1] = -e_ip    - z*e_mK2
    B[3, 3] =  E
    B[3, 8] = -1       - z*e_mK2

    # (5)  E C21 − C23 − e^{-iφ}C11 − e^{iφ}C22
    #      − z^{-1}(e^{iK/2}C11 + e^{-iK/2}(C22+C32)) = 0
    B[4, 0] = -e_imp - zinv*e_iK2
    B[4, 1] = -e_ip  - zinv*e_mK2
    B[4, 4] =  E
    B[4, 7] = -1     - zinv*e_mK2

    # (6)  E C13 − C11 − C33 − e^{iφ}C23
    #      − z(e^{iK/2}C11 + e^{-iK/2}(C23+C33)) = 0
    B[5, 0] = -1 - z*e_iK2
    B[5, 2] = -1 - z*e_mK2
    B[5, 5] =  E
    B[5, 7] = -e_ip - z*e_mK2

    # (7)  E C31 − C11 − C33 − e^{iφ}C32
    #      − z^{-1}(e^{iK/2}C11 + e^{-iK/2}(C32+C33)) = 0
    B[6, 0] = -1 - zinv*e_iK2
    B[6, 2] = -1 - zinv*e_mK2
    B[6, 6] =  E
    B[6, 8] = -e_ip - zinv*e_mK2

    # (8)  E C23 − C21 − e^{-iφ}C13 − z^{-1}e^{iK/2}C13 − z e^{iK/2}C21 = 0
    B[7, 4] = -1     - z*e_iK2
    B[7, 5] = -e_imp - zinv*e_iK2
    B[7, 7] =  E

    # (9)  E C32 − C12 − e^{-iφ}C31 − z^{-1}e^{iK/2}C12 − z e^{iK/2}C31 = 0
    B[8, 3] = -1     - zinv*e_iK2
    B[8, 6] = -e_imp - z*e_iK2
    B[8, 8] =  E

    return B


# ----------------------------------------------------------------------
# helper: build the 9×9 BOUNDARY+SYMMETRY matrix M(E,z;K,phi,U)
# rows: 6 boundary eqns  + 3 symmetry (C21=C12 etc.)
# ----------------------------------------------------------------------
def boundary_matrix(E, z, K, phi, U):
    e_ip,  e_imp  = exp(1j*phi),  exp(-1j*phi)
    e_iK2, e_mK2  = exp(1j*K/2),  exp(-1j*K/2)
    cK2, zinv     = cos(K/2), 1/z

    M = np.zeros((9, 9), dtype=complex)

    # shorthand indices
    C11,C22,C33,C12,C21,C13,C31,C23,C32 = range(9)

    # --- boundary m=0 (rows 0–5) ----------------------------------
    M[0, C11] = E - U
    M[0, C12] = -2*e_ip - 2*cK2*zinv
    M[0, C13] = -2       - 2*cK2*zinv

    M[1, C22] = E - U
    M[1, C12] = -2*e_imp - 2*cK2*zinv
    M[1, C13] = -2

    M[2, C33] = E - U
    M[2, C13] = -4 - 2*cK2*zinv

    M[3, C11] = -e_imp - zinv*e_iK2
    M[3, C22] = -e_ip  - zinv*e_mK2
    M[3, C12] =  E
    M[3, C13] = -1
    M[3, C23] = - zinv*e_mK2

    M[4, C11] = -1 - zinv*e_iK2
    M[4, C33] = -1 - zinv*e_mK2
    M[4, C13] =  E
    M[4, C23] = -e_ip - zinv*e_mK2

    M[5, C12] = - zinv*e_iK2
    M[5, C13] = -e_imp
    M[5, C23] =  E - 1 - zinv*e_iK2

    # --- bosonic symmetry at m=0 (rows 6–8) -----------------------
    #  C21 - C12 = 0 ,  C31 - C13 = 0 ,  C32 - C23 = 0
    M[6, C21] =  1;  M[6, C12] = -1
    M[7, C31] =  1;  M[7, C13] = -1
    M[8, C32] =  1;  M[8, C23] = -1

    return M


# ----------------------------------------------------------------------
# residual vector for least‑squares: 4 real equations
# ----------------------------------------------------------------------
def residual(vec, K, phi, U):
    E, z = vec
    # keep solver inside physical region
    if z <= 0.0 or z >= 1.0:
        return np.array([1e3, 1e3, 1e3, 1e3])

    detB = np.linalg.det(bulk_matrix(E, z, K, phi))
    detM = np.linalg.det(boundary_matrix(E, z, K, phi, U))

    return np.array([detB.real, detB.imag, detM.real, detM.imag])


# ----------------------------------------------------------------------
# driver: solve for a single K using least_squares
# ----------------------------------------------------------------------
def solve_one_K(K, phi, U, guess):
    sol = least_squares(
        residual, guess,
        args=(K, phi, U),
        method='trf',
        xtol=1e-12, ftol=1e-12, gtol=1e-12,
        max_nfev=400
    )
    if sol.cost > 1e-24 or not sol.success:
        return None
    E, z = sol.x
    return E, z


# ----------------------------------------------------------------------
# scan the Brillouin zone
# ----------------------------------------------------------------------
def dispersion(U, phi, NK=41):
    K_vals  = np.linspace(-np.pi, np.pi, NK)
    E_band  = np.full(NK, np.nan)
    z_vals  = np.full(NK, np.nan)

    # heuristic initial guess
    guess = np.array([U + 1.0 if U > 0 else U - 1.0, 0.3])

    for i, K in enumerate(K_vals):
        sol = solve_one_K(K, phi, U, guess)
        if sol is None:
            print(f"[warn] failed at K={K:.3f}")
            continue
        E_band[i], z_vals[i] = sol
        guess = sol  # continuation along K

    return K_vals, E_band, z_vals


# ----------------------------------------------------------------------
# quick demo
# ----------------------------------------------------------------------
if __name__ == "__main__":
    U   = -5.0          # interaction
    phi = 0       # Peierls phase
    NK  = 11            # K grid

    K_vals, E_band, z_vals = dispersion(U, phi, NK)

    plt.plot(K_vals, E_band, 'o-', ms=3)
    plt.xlabel(r'$K$')
    plt.ylabel(r'$E_{\mathrm{dimer}}(K)$')
    plt.title(fr'Dimer dispersion  ($U={U}$, $\phi={phi:.2f}$)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
